```python
import scipy.integrate as integrate
import numpy as np
import matplotlib.pyplot as plt

T=[]
l=0.2
g=9.81
def K(x,ang_ini):     #Funci√≥n K(k)
    k=np.sin(np.deg2rad(ang_ini)/2)
    f=1/np.sqrt(1-(k**2)*(np.sin(x))**2)
    return(f)

ang_ini=np.linspace(0,60,7)  #Angulos iniciales 

for i in range(len(ang_ini)):
    integral=integrate.quad(K, 0, 2*np.pi, args=ang_ini[i] ) #Integral de K(k) con diferentes √°ngulos iniciales
    T.append(4*np.sqrt(l/g)*(integral[0])) #Se saca s√≥lo el valor de la integral que est√° en la posici√≥n 0 de la lista que entrega integrate
print(T)

plt.plot(ang_ini, T, 'm-')
plt.grid()
plt.title('T vs. √Ångulo inicial')
plt.ylabel('T')
plt.xlabel('√Ångulo inicial')
    
```
