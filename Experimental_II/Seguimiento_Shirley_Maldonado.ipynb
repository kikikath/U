```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

##### Campo magnetico de electrodomesticos

```python
#Celular off
celular_off=pd.read_excel('Celular apagado.xlsx')    #Para leer el archivo s칩lo debe descargarlo en el pc
cel_off = celular_off['BT'][0:1001]
t_cof = celular_off['time'][0:1001]
N_cof = len(cel_off)
t = np.arange(0,t_cof[1000],1)
print('Promedio Celular off=', np.mean(cel_off))
print('Desviaci칩n Celular off=', np.std(cel_off))
print('Error est치ndar celular off=', np.std(cel_off)/np.sqrt(N_cof), '\n')

#Celular on
celular_on=pd.read_excel('Celular encendido.xlsx')
cel_on = celular_on['BT'][0:1001]
t_con = celular_on['time'][0:1001]
N_con = len(cel_on)
t = np.arange(0,t_con[1000],1)
print('Promedio Celular on=', np.mean(cel_on))
print('Desviaci칩n Celular on=', np.std(cel_on))
print('Error est치ndar celular on=', np.std(cel_on)/np.sqrt(N_con), '\n')

plt.figure(figsize=(12,3))
plt.plot(t_cof, cel_off, lw=1, color='c', ms=6, markerfacecolor='blue', label='Campo magnetico Cel off')
plt.plot(t_con, cel_on, lw=1, color='r', ms=6, markerfacecolor='r', label='Campo magnetico Cel on')

plt.title('Campo magnetico del celular en diferentes estados vs. Tiempo')
plt.ylabel('Campo magentico (췃T)',fontsize=12)
plt.xlabel('t(s) \n Figura 1.',fontsize=12)
plt.legend()
plt.show()
```

```python
#Impresora des
imp_des=pd.read_excel('Impresora desconectada .xlsx')    #Para leer el archivo s칩lo debe descargarlo en el pc
impr_des = imp_des['BT'][0:800]
t_impd = imp_des['time'][0:800]
N_impd = len(impr_des)
t = np.arange(0,t_impd[799],1)
print('Promedio impresora des=', np.mean(impr_des))
print('Desviaci칩n impresora des=', np.std(impr_des))
print('Error est치ndar impresora des=', np.std(impr_des)/np.sqrt(N_impd), '\n')

#Impresora con
imp_con = pd.read_excel('Impresora conectada .xlsx')
im_con = imp_con['BT'][0:800]
t_imc = imp_con['time'][0:800]
N_imc = len(im_con)
t = np.arange(0,t_imc[799],1)
print('Promedio impresora con=', np.mean(im_con))
print('Desviaci칩n impresora con=', np.std(im_con))
print('Error est치ndar impresora con=', np.std(im_con)/np.sqrt(N_imc), '\n')

#Impresora On
imp_on = pd.read_excel('Impresora funcionando .xlsx')
im_on = imp_on['BT'][0:800]
t_imo = imp_on['time'][0:800]
N_imo = len(im_on)
t = np.arange(0,t_imo[799],1)
print('Promedio impresora on=', np.mean(im_on))
print('Desviaci칩n impresora on=', np.std(im_on))
print('Error est치ndar impresora on=', np.std(im_on)/np.sqrt(N_imo))


plt.figure(figsize=(12,5))
plt.plot(t_impd, impr_des, lw=1, color='c', ms=6, label='Campo magnetico impresora des')
plt.plot(t_imc, im_con, lw=1, color='r', ms=6, label='Campo magnetico impresora con')
plt.plot(t_imo, im_on, lw=1, color='g', ms=6, label='Campo magnetico impresora on')

plt.title('Campo magnetico de una impresora en diferentes estados vs. Tiempo')
plt.ylabel('Campo magentico (췃T)',fontsize=12)
plt.xlabel('t(s) \n Figura 2.',fontsize=12)
plt.legend()
plt.show()
```

##### 6. Sonidos 

```python
#Sonido 1
S1=pd.read_excel('Sonido 1.xlsx')   #Tuve que convertilo a formato excel porque con el formato csv s칩lo mostraba 1 columna.
Ns1=len(S1)
s1 = S1['gain'][0:Ns1] 
print('SONIDO 1')

plt.figure(figsize=(10,5))
n, bins, x = plt.hist(s1, 100, color='gold', alpha=1, edgecolor='black') #Gr치fica de histograma
plt.xlabel('Campo magnetico (췃T) \n \n Figura 4. Histograma de sonido 1')
plt.ylabel('Frecuencia')

m_s1 = np.mean(s1)
des_s1 = np.std(s1)
er_s1 = des_s1 / np.sqrt(Ns1)
y = norm.pdf( bins, m_s1, des_s1)
plt.plot(bins, y*Ns1*abs(bins[0]-bins[1]), 'r', linewidth=1, label='Curva de ajuste')
plt.legend()
plt.grid()
plt.show()

print('N=', Ns1)
print('Media=',m_s1)
print('Desviaci칩n est치ndar=', des_s1)
print('Error est치ndar=', er_s1)
er_ers1 = er_s1*(7.1/100)
print('Error en el error:', er_ers1)
print('Valor medio: (41 췀 5) 췃T')
print('Valor medio con error: (41 췀 0.2) 췃T\n')


#Sonido 2
S2=pd.read_excel('Sonido 2.xlsx')   
Ns2=len(S2)
s2 = S2['gain'][0:Ns2] 
print('SONIDO 2')

plt.figure(figsize=(10,5))
n, bins, x = plt.hist(s2, 100, color='gold', alpha=1, edgecolor='black') #Gr치fica de histograma
plt.xlabel('Campo magnetico (췃T) \n \n Figura 5. Histograma de sonido 2')
plt.ylabel('Frecuencia')

m_s2 = np.mean(s2)
des_s2 = np.std(s2)
er_s2 = des_s2 / np.sqrt(Ns2)
y = norm.pdf( bins, m_s2, des_s2)

plt.plot(bins, y*Ns2*abs(bins[0]-bins[1]), 'r', linewidth=1, label='Curva de ajuste')
plt.legend()
plt.grid()
plt.show()

print('N=', Ns2)
print('Media=',m_s2)
print('Desviaci칩n est치ndar=', des_s2)
print('Error est치ndar=', er_s2)
er_ers2 = er_s2*(7.1/100)
print('Error en el error:', er_ers2)
print('Valor medio: (52 췀 4) 췃T')
print('Valor medio: (52 췀 0.2) 췃T\n')

#Sonido 3
S3=pd.read_excel('Sonido 3.xlsx')  
Ns3=len(S3)
s3 = S3['gain'][0:Ns3]
print('SONIDO 3')

plt.figure(figsize=(10,5))
n, bins, x = plt.hist(s3, 100, color='gold', alpha=1, edgecolor='black') #Gr치fica de histograma
plt.xlabel('Campo magnetico (췃T) \n \n Figura 6. Histograma de sonido 3')
plt.ylabel('Frecuencia')

m_s3 = np.mean(s3)
des_s3 = np.std(s3)
er_s3 = des_s3 / np.sqrt(Ns3)
y = norm.pdf( bins, m_s3, des_s3) 
plt.plot(bins, y*(Ns3)*abs(bins[0]-bins[1]), 'r-', linewidth=1, label='Curva de ajuste')
plt.legend()
plt.grid()
plt.show()

print('N=', Ns3)
print('Media=',m_s3)
print('Desviaci칩n est치ndar=', des_s3)
print('Error est치ndar=', er_s3)
er_ers3 = er_s3*(7.1/100)
print('Error en el error:', er_ers3)
print('Valor medio: (50 췀 5) ')
print('Valor medio: (50 췀 0.2) 췃T')
```

```python
#Sonido 1
nn = 5
bN = int(np.sqrt(len(s1)/nn))

s11=[]
for ii in range(int(len(s1)/nn)):
  prome = np.sum(s1[int(ii*nn+1):int((ii+1)*nn+1)])/nn
  s11.append(prome)

plt.figure(figsize=(10,5))
n, bins, x = plt.hist(s11, 50, color='gold', edgecolor='black') #Gr치fica de histograma
plt.xlabel('Campo magnetico (췃T) \n \n Figura 7. Histograma de sonido 1')
plt.ylabel('Frecuencia')
plt.title('(41 췀 4) 췃T')

m_s11 = np.mean(s11)
des_s11 = np.std(s11)
er_s11 = des_s11 / np.sqrt(len(s11))
y = norm.pdf( bins, m_s11, des_s11)
plt.plot(bins, y*len(s11)*abs(bins[0]-bins[1]), 'r', label='Curva de ajuste')
plt.legend()
plt.grid()
plt.show()


print('SONIDO 1')
print('Tama침o de la muestra = %d'%(int(len(s1)/nn)))
print('Media=',m_s11)
print('Desviaci칩n est치ndar=', des_s11)
print('Error est치ndar=', er_s11)
er_ers11 = er_s11*(7.1/100)
print('Error en el error:', er_ers11)
print('Valor reportado: 41 +- 4\n')
```
