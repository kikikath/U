```python
x=3
y=4
suma=x+y
resta=x-y
multiplicacion=x*y
division=x/y
print ("la suma entre", x, "y", y, "es:",suma)
print ("la resta entre", x, "y", y, "es:",resta)
print ("la multiplicación entre", x, "y", y, "es:",multiplicacion)
print ("la división entre", x, "y", y, "es:",division)
```
