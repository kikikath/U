```python
import rebound as rb
import matplotlib.pyplot as plt
from celluloid import Camera
from IPython.core.display import HTML
import numpy as np
import pandas as pd
```

```python
sist = rb.Simulation()
sist.add(m=0.5,x=0,y=0,z=0,vx=0.5,vy=0.5,vz=0)
sist.add(m=0.5,x=1,y=0,z=0,vx=0.5,vy=-1,vz=0)
sist.add(m=0.5,x=1,y=1,z=0,vx=-1,vy=0,vz=0)
sist.add(m=0.5,x=0,y=1,z=0,vx=0.5,vy=-0.5,vz=0)
```

```python
fig=plt.figure()
cam = Camera(fig)

for t in range(50):
  sist.integrate(t)

  plt.plot(sist.particles[0].x,sist.particles[0].y,'co')
  plt.plot(sist.particles[1].x,sist.particles[1].y,'ro')
  plt.plot(sist.particles[2].x,sist.particles[2].y,'bo')
  plt.plot(sist.particles[3].x,sist.particles[3].y,'mo')

  cam.snap()

plt.close()
```

```python
animacion=cam.animate()
```

```python
P_p1 = []
P_p2 = []
P_p3 = []
P_p4 = []
Px_t = []
Py_t = []
Pz_t = []
P_total = []
L_p1 = []
L_p2 = []
L_p3 = []
L_p4 = []
L_total = []
for t in range(50):
  sist.integrate(t)

  # Calculo de la magnitud de la velocidad
  vel_1 = np.sqrt( sist.particles[0].vx**2 + sist.particles[0].vy**2 + sist.particles[0].vz**2)
  vel_2 = np.sqrt( sist.particles[1].vx**2 + sist.particles[1].vy**2 + sist.particles[1].vz**2)
  vel_3 = np.sqrt( sist.particles[2].vx**2 + sist.particles[2].vy**2 + sist.particles[2].vz**2)
  vel_4 = np.sqrt( sist.particles[3].vx**2 + sist.particles[3].vy**2 + sist.particles[3].vz**2)

  # Cálculo de la magnitud del momentum lineal
  p1 = sist.particles[0].m * vel_1
  p2 = sist.particles[1].m * vel_2
  p3 = sist.particles[2].m * vel_3
  p4 = sist.particles[3].m * vel_4
  pTotal = p1+p2+p3+p4

  # Calculo del momentum lineal total por componentes
  px1 = sist.particles[0].m * sist.particles[0].vx
  py1 = sist.particles[0].m * sist.particles[0].vy
  pz1 = sist.particles[0].m * sist.particles[0].vz
  px2 = sist.particles[1].m * sist.particles[1].vx
  py2 = sist.particles[1].m * sist.particles[1].vy
  pz2 = sist.particles[1].m * sist.particles[1].vz
  px3 = sist.particles[2].m * sist.particles[2].vx
  py3 = sist.particles[2].m * sist.particles[2].vy
  pz3 = sist.particles[2].m * sist.particles[2].vz
  px4 = sist.particles[3].m * sist.particles[3].vx
  py4 = sist.particles[3].m * sist.particles[3].vy
  pz4 = sist.particles[3].m * sist.particles[3].vz

  Px_t.append(px1 + px2 + px3 + px4)
  Py_t.append(py1 + py2 + py3 + py4)
  Pz_t.append(pz1 + pz2 + pz3 + pz4)

  P_p1.append(np.round(p1, 4))
  P_p2.append(np.round(p2, 4))
  P_p3.append(np.round(p3, 4))
  P_p4.append(np.round(p4, 4))
  P_total.append(np.round(pTotal, 4))

  # Cálculo del momentum angular
  #Como las componentes de la posición y la velocidad en z son 0, entonces, el producto
  #cruz entre r y p, o sea, el vector L, queda en la dirección z. 
  L1 = (sist.particles[0].x * sist.particles[0].m * sist.particles[0].vy) - (sist.particles[0].y * sist.particles[0].m * sist.particles[0].vx)
  L2 = (sist.particles[1].x * sist.particles[1].m * sist.particles[1].vy) - (sist.particles[1].y * sist.particles[1].m * sist.particles[1].vx)
  L3 = (sist.particles[2].x * sist.particles[2].m * sist.particles[2].vy) - (sist.particles[2].y * sist.particles[2].m * sist.particles[2].vx)
  L4 = (sist.particles[3].x * sist.particles[3].m * sist.particles[3].vy) - (sist.particles[3].y * sist.particles[3].m * sist.particles[3].vx)
  Ltotal = L1+L2+L3+L4
  L_p1.append(np.round(L1,4))
  L_p2.append(np.round(L2,4))
  L_p3.append(np.round(L3,4))
  L_p4.append(np.round(L4,4))
  L_total.append(np.round(Ltotal,4))
```

```python
df = pd.DataFrame()

df['P part 1'] = P_p1
df['P part 2'] = P_p2
df['P partí 3 '] = P_p3
df['P part 4'] = P_p4
df['P total en x'] = Px_t
df['P total en y'] = Py_t
df['P total en z'] = Pz_t
df['P total'] = P_total

df['L part 1'] = L_p1
df['L part 2'] = L_p2
df['L part 3'] = L_p3
df['L part 4'] = L_p4

df['L total'] =L_total
print(df)
```

**Sobre el momentum lineal:** Es posible ver que la magnitud del momentum lineal total varía durante el tiempo, esto se debe a que la magnitud del momentum total es la suma de las magnitudes del momentum de cada partícula, y estás magnitudes no muestran la dirección ni el sentido del momentum.

Mientras que si vemos el momentum lineal total por componentes, este se conserva. De esta forma, se puede concluir que el momentum lineal total se conserva pues el cambio de la magnitud total, se debe al no especificar las direcciones del momentum angular de cada partícula al sumarlas.



**Sobre el momentum angular:** El resultado anterior del momentum angular total se dio con un signo negativo, esto se debe a que la magnitud es la dada (0.25) pero el signo representa la dirección a la que apunta el vector del momentum angular, es decir, el vector del momentum angular apunta a la dirección negativa del vector unitario k.

Para el caso del momentum angular los resultados dados sí se conserva pues al hallar el momentum angular de cada partícula, el sentido del vector se mantiene (o sea, el signo) por lo que al sumarlos todo, el momentum angular total también presenta el sentido del vector de momentum angular total.
