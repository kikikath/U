### Semieje mayor - histograma


```python
def lista(a, b):
    aa = []
    bb = []
    for i in range(len(a)):
        k = a[i]
        j = b[i]
        if k != 0:
            aa.append(k)
            bb.append(j)
    return aa, bb

i, a = lista(np.array(df['inclination']), np.array(df['semi_major_axis']))
print("Número de exoplanetas con semiejes mayores", len(a))
a.remove(np.max(a))
a.remove(np.max(a))
a.remove(np.max(a))
a.remove(np.max(a))
a.remove(np.max(a))
a.remove(np.max(a))
a.remove(np.max(a))
a.remove(np.max(a))
a.remove(np.max(a))
a.remove(np.max(a))
a.remove(np.max(a))
a.remove(np.max(a))
a.remove(np.max(a))
a.remove(np.max(a))
```

```python
intervalos = range(int(np.min(a)), int(np.max(a))) #calculamos los extremos de los intervalos

inter = np.linspace(int(np.min(a)), int(np.max(a)), 50000)
```

```python
plt.hist(x=a, bins=inter, color='m')
plt.ylabel('Frecuencia')
plt.xlabel('Semieje mayor [AU]')
plt.title('Histograma del semieje mayor de exoplanetas')
plt.xlim(0,1)
plt.grid()
plt.show()
```
