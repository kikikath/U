**1** Realice una función que elija dos numeros `a` y `b` de forma aleatoria varias veces y realice solo una operación de forma aleatoria (suma, resta,divición o multiplicación) con estos valores, a la vez guarde los resultados en un diccionario y finalmente escriba esto en un archivo y guardelo.

Ejemplo del return:

>Numero de iteraiónes: 2
>>a=10,2,b=5,5, Suma=1,0,Multiplicaión=0,1,resta=0,0,Divición=0,0,resultado=15,10

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def opera():
    N=int(input("Ingrese número entero de iteraciones mayor a 0:")) 
    D={}                                      #Crear diccionario
    La=[]                                     #Lista para a
    Lb=[]                                     #Lista para b
    re=[]                                     #Lista para el resultado
    c=random.random()
    
    if c<=0.25:                               #Condicional para cada operación
        op="Suma="
        for i in range (1,N+1,1):
                a=random.randint(0,10)        #Formar a y b
                b=random.randint(0,10)
                La.append(a)                  #Añadir a y b a la lista
                Lb.append(b)
                k= a+b                        #Hacer operación 
                re.append(k)                  #Añadir a la lista

                
    elif 0.25<c<=0.5:    
        op="Resta="
        for i in range (1,N+1,1):
                a=random.randint(0,10)
                b=random.randint(0,10)
                La.append(a)
                Lb.append(b)
                k= a-b 
                re.append(k)
                
                
    elif 0.5<c<=0.75:   
        op="Multiplicacion="
        for i in range (1,N+1,1):
                a=random.randint(0,10)
                b=random.randint(0,10)
                La.append(a)
                Lb.append(b)
                k= a*b 
                re.append(k)
                
    else:       
        op="Division="
        for i in range (1,N+1,1):
                a=random.randint(0,10)
                b=random.randint(0,10)
                La.append(a)
                Lb.append(b)
                if b==0:
                    k="indef"
                else:
                    k=a/b
                re.append(k)

    D["a="] = La                              #Añadir cada lista al diccionario
    D["b="] = Lb
    D['{}'.format(op)] = re
    return(D)                                 #Fin de la función

D_O = opera()                                 
print(D_O)

ar=pd.DataFrame()                               #Crear el archivo y guardar
ar=ar.append(D_O, ignore_index=True)        
ar.to_excel('Operación_aleatoria.xlsx', index=False) 
```

**2** Realice el problema del borracho visto en clase en dos dimensiomnes, es decir, que el desplazamiento se de tanto en `x` como en `y` y grafique su desplazamiento. Recuerde que el paso se puede dar en x o en y de forma aleatoria.

Ejemplo:



```python
k=2000 #(Numero de pasos)

def borr_2d():
    m=0 #punto de incio en x
    n=0 #punto de inicio en y
    M=[]
    N=[]
    for i in range (1,k+1):
        hv= random.random() 
        if hv<=0.25:           #paso hacia arriba
            m=m+1
        elif 0.25<hv<=0.5:     #paso hacia abajo
            m=m-1
        elif 0.5<hv<=0.75:     #paso a la derecha
             n=n+1
        else:          #paso a la izquierda
             n=n-1 
        M.append(m)
        N.append(n)
        
    return(M,N)
M,N=borr_2d()
plt.plot(N,M,linewidth=1, color="gold" )
```

**3** Ahora realice nuevamente el problema del borracho pero esta vez sin restricciones en la dirreccion, es decir, que el haya un ángulo asociado, grafique.

Ejemplo:



```python
k=740 #(Numero de pasos)


h=0 #punto de incio en x
v=0 #punto de inicio en y
H=[]
V=[]
for i in range (1,k+1):
    hv= random.uniform(0,2*np.pi) #Ángulo aleatorio
    h=np.sin(hv)+h                #Se usa las funciones trigonometricas para hallar "x" y "y"
    v=np.cos(hv)+v                #tomando 1 como la magnitud de cada paso
    H.append(h)
    V.append(v)

plt.plot(H,V,'k', linewidth=1, color="magenta" )


plt.xlim( min(H), max(H) ) #Se ajustan los limites para que la gráfica quede enfocada en el movimiento
plt.ylim( min(V), max(V) )

plt.minorticks_on() 
```
